```python
import hopsworks

project = hopsworks.login()

fs = project.get_feature_store()
```

```python
city_weather_fg = fs.get_feature_group('city_weather', version=1)
query = city_weather_fg.select_all()
df_city_weather=query.read()
```

```python
df_city_weather
```

```python
df_city_weather.drop(columns=['index_column'],inplace=True)
```

```python
#Identify and drop duplicate rows weather : [city_id,date,hour]
# Identify and drop duplicate rows based on 'city_id', 'date', and 'hour' columns
df_city_weather_cleaned = df_city_weather.drop_duplicates(subset=['city_id', 'datetime'])

# Optionally, reset the index after dropping duplicates
df_city_weather_cleaned.reset_index(drop=True, inplace=True)

# Display the cleaned DataFrame
print(df_city_weather_cleaned.head())
```

```python
df_city_weather_cleaned = df_city_weather_cleaned.copy()
df_city_weather_cleaned.drop(columns=['chanceofrain','chanceoffog','chanceofthunder','chanceofsnow'],inplace=True)
```

```python

df_city_weather_cleaned.drop(columns=['event_time'],inplace=True)
```

```python
df_city_weather_cleaned
```
