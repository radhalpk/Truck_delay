```python
# Convert naive timestamps to UTC
train_cutoff = pd.to_datetime('2019-01-30').tz_localize('UTC')
validation_cutoff = pd.to_datetime('2019-02-07').tz_localize('UTC')
```

```python
# 3. Splitting the data into training, validation, and test sets based on estimated_arrival
train_df = final_df[final_df['estimated_arrival'] <= train_cutoff]
validation_df = final_df[(final_df['estimated_arrival'] > train_cutoff) & 
                         (final_df['estimated_arrival'] <= validation_cutoff)]
test_df = final_df[final_df['estimated_arrival'] > validation_cutoff]
```

```python
print("Columns in train_df:", train_df.columns.tolist())
print("Columns in validation_df:", validation_df.columns.tolist())
print("Columns in test_df:", test_df.columns.tolist())
```

```python
# Define your continuous columns (based on available columns)
cts_cols = [
    'route_avg_temp', 'route_avg_wind_speed', 'route_avg_precip', 'route_avg_humidity',
    'route_avg_visibility', 'route_avg_pressure', 'distance', 'average_hours',
    'avg_no_of_vehicles', 'truck_age', 'load_capacity_pounds', 'mileage_mpg',
    'age', 'experience', 'average_speed_mph'
]

# Define your categorical columns based on available columns in train_df
cat_cols = [
    'route_description',  # This is available
    'description_origin',  # Changed from 'origin_description' to 'description_origin'
    'description_destination',  # Changed from 'destination_description' to 'description_destination'
    'accident',  # This is available
    'fuel_type',  # This is available
    'gender',  # This is available
    'driving_style',  # This is available
    
]

target = ['delay']

# Proceed to separate features and target variables
# Prepare features and target variables for training, validation, and test sets
X_train = train_df[cts_cols + cat_cols]
y_train = train_df['delay']

X_valid = validation_df[cts_cols + cat_cols]
y_valid = validation_df['delay']

X_test = test_df[cts_cols + cat_cols]
y_test = test_df['delay']
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
import pandas as pd

# Step 1: OneHotEncoding the categorical columns
encoder = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')

# Fit the encoder on the training data
X_train_encoded = encoder.fit_transform(X_train[cat_cols])
X_valid_encoded = encoder.transform(X_valid[cat_cols])
X_test_encoded = encoder.transform(X_test[cat_cols])

# Generating feature names for the new one-hot encoded features
encoded_features = encoder.get_feature_names_out(cat_cols)

# Convert the encoded arrays back to DataFrames using the original index
X_train_encoded_df = pd.DataFrame(X_train_encoded, columns=encoded_features, index=X_train.index)
X_valid_encoded_df = pd.DataFrame(X_valid_encoded, columns=encoded_features, index=X_valid.index)
X_test_encoded_df = pd.DataFrame(X_test_encoded, columns=encoded_features, index=X_test.index)

# Drop the original categorical features from X_train, X_valid, and X_test, and join the encoded features
X_train_final = X_train.drop(cat_cols, axis=1).join(X_train_encoded_df)
X_valid_final = X_valid.drop(cat_cols, axis=1).join(X_valid_encoded_df)
X_test_final = X_test.drop(cat_cols, axis=1).join(X_test_encoded_df)

# Step 2: Apply SMOTE after encoding (only to training data)
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_final, y_train)

# Step 3: Apply StandardScaler after SMOTE
scaler = StandardScaler()

# Fit the scaler on the resampled training data and transform the train, validation, and test data
X_train_scaled = scaler.fit_transform(X_train_resampled)
X_valid_scaled = scaler.transform(X_valid_final)
X_test_scaled = scaler.transform(X_test_final)

# Optionally, convert the scaled arrays back to DataFrames using the original index and columns
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train_final.columns)
X_valid_scaled_df = pd.DataFrame(X_valid_scaled, columns=X_valid_final.columns, index=X_valid_final.index)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test_final.columns, index=X_test_final.index)

# Check the transformed data (optional)
print(X_train_scaled_df.head())
```

```python
import mlflow
import mlflow.sklearn
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Set up the experiment name
mlflow.set_experiment("Truck_Delay_Prediction")

# Define a function to train and log the model
def train_and_log_model(model_name, grid_search, X_train, y_train, X_valid, y_valid, X_test, y_test):
    # End any existing MLflow run
    mlflow.end_run()
    
    with mlflow.start_run(run_name=model_name):
        # Fit the model using GridSearchCV
        grid_search.fit(X_train, y_train)
        
        # Get the best estimator
        best_model = grid_search.best_estimator_
        
        # Log best parameters
        mlflow.log_params(grid_search.best_params_)
        
        # Predict on training, validation, and test sets
        y_train_pred = best_model.predict(X_train)
        y_valid_pred = best_model.predict(X_valid)
        y_test_pred = best_model.predict(X_test)
        
        # Calculate metrics
        metrics = {}
        # Training metrics
        metrics['train_accuracy'] = accuracy_score(y_train, y_train_pred)
        metrics['train_precision'] = precision_score(y_train, y_train_pred, average='weighted')
        metrics['train_recall'] = recall_score(y_train, y_train_pred, average='weighted')
        metrics['train_f1_score'] = f1_score(y_train, y_train_pred, average='weighted')
        # Validation metrics
        metrics['valid_accuracy'] = accuracy_score(y_valid, y_valid_pred)
        metrics['valid_precision'] = precision_score(y_valid, y_valid_pred, average='weighted')
        metrics['valid_recall'] = recall_score(y_valid, y_valid_pred, average='weighted')
        metrics['valid_f1_score'] = f1_score(y_valid, y_valid_pred, average='weighted')
        # Test metrics
        metrics['test_accuracy'] = accuracy_score(y_test, y_test_pred)
        metrics['test_precision'] = precision_score(y_test, y_test_pred, average='weighted')
        metrics['test_recall'] = recall_score(y_test, y_test_pred, average='weighted')
        metrics['test_f1_score'] = f1_score(y_test, y_test_pred, average='weighted')
        
        # Log metrics
        mlflow.log_metrics(metrics)
        
        # Log the model
        mlflow.sklearn.log_model(best_model, model_name)
        
        # Print metrics
        print(f"\nModel: {model_name}")
        print(f"Best Parameters: {grid_search.best_params_}")
        print(f"Training Accuracy: {metrics['train_accuracy']:.4f}")
        print(f"Validation Accuracy: {metrics['valid_accuracy']:.4f}")
        print(f"Test Accuracy: {metrics['test_accuracy']:.4f}")
        print(f"Training F1 Score: {metrics['train_f1_score']:.4f}")
        print(f"Validation F1 Score: {metrics['valid_f1_score']:.4f}")
        print(f"Test F1 Score: {metrics['test_f1_score']:.4f}")
```
